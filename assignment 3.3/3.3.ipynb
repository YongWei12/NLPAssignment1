```python
import nltk
import pandas as pd
import operator

from nltk.tokenize import word_tokenize
from nltk.tokenize import TweetTokenizer, sent_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#import data
data = pd.read_csv(r'.\data\DataSet.csv')
df = data['reviews__review']
df[0]
```

```python
# Main idea
#for each review:
#     tokenize each sentence and analyse
#     sum is it positive or negative for each review
# sum up all the review for 1 product
```

```python
#initialise
tokenizer_words = TweetTokenizer()
sentence = []
token_sentence = []
```

```python
#tokenize sentence
for i in range(0,len(df)):
    sentence.append(nltk.sent_tokenize(df[i]))
```

```python
sentence[0]
```

```python
#sentence analyser
sid = SentimentIntensityAnalyzer()
```

```python
#initialise for score
sentence_score = []
review_score = []
```

```python
#compute score

for i in range(0,len(sentence)): #per review
    sentence_score = []
    for j in range(0,len(sentence[i])): #for each sentence in each review
        ss = sid.polarity_scores(sentence[i][j])
        print(sentence[i][j])
        print( f"Compound score: {ss['compound']}")
        print()
        sentence_score.append(ss['compound'])
    review_score.append(sum(sentence_score))

#final score for that 1 product
product_score = sum(review_score)
```
